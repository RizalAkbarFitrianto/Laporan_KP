```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt



train_df = pd.read_csv('dataset_dakota.txt')
train_df.head()
```

```python
# Scatterplot
sns.scatterplot(data=train_df,x='CUSTOMER',y='PENDAPATAN')
plt.title('Hubungan antara Customer dan Pendapatan')
```

```python
# Lineplot
plt.plot(train_df['BULAN'], train_df['PENDAPATAN'])
```

```python
# Barplot (Horizontal)
sns.barplot(data=train_df,x="BULAN",y="PENDAPATAN")
```

```python
# Heatmap
data = train_df
korelasi = data.corr()
sns.heatmap(korelasi, annot=True, cmap='coolwarm')
plt.show()
```
